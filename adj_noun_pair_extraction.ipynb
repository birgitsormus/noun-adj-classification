### Creating pairs from phrases

Creating meaningful pairs from the output received from GrammarTagger.

```python
paarid = []
ühendid = text.ühendid

for ühend in ühendid:
    if len(ühend) == 2:
        if ühend[1].partofspeech[0] == 'J' or ühend[1].partofspeech[0] == 'Z':
            continue
        paarid.append((ühend[0].lemma[0], ühend[1].lemma[0]))
    elif len(ühend) == 3:
        teine = ühend[1].lemma[0]
        is_verb = True if ühend[1].partofspeech[0] == 'V' else False
        esimene = ühend[0].lemma[0]
        kolmas = ühend[2].lemma[0]
        if is_verb:
            paarid.append((kolmas, esimene))
        else:
            paarid.append((esimene, kolmas))
            paarid.append((teine, kolmas))
    else:
        noun = ühend[-1].lemma[0]
        for sõna in ühend:
            if sõna.partofspeech[0] == 'A':
                paarid.append((sõna.lemma[0], noun))
```

```python
# Create adjective and noun pair table
try:
    pair_table = "pair_counts"
    create_table_query = '''CREATE TABLE '''+ pair_table +''' 
              (adjective      TEXT    NOT NULL,
               noun           TEXT    NOT NULL,
               count          INT     NOT NULL   
               ); '''

    #Execute this command (SQL Query)
    cursor.execute(create_table_query)
    
    # Make the changes to the database persistent
    con.commit()
    print("Table ("+ pair_table +") created successfully in PostgreSQL ")
except (Exception, psycopg2.Error) as error:
    # if it exits with an exception the transaction is rolled back.
    con.rollback()
    print("Error while creating the table: ", error)
```
